```python
#import library yang dibutuhkan
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
import cv2

jokowi_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/jokowi.jpg')
jokowi_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/jokowi.jpg')

kartini_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/kartini.jpg')
kartini_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/kartini.jpg')

kucing_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Images/facedet/kucing.jpg')
kucing_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Images/facedet/kucing.jpg')

mask_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/mask.png')
mask_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/mask.png')

jordan_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/mjordan.jpg')
jordan_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/mjordan.jpg')

solvayconf_img1 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/solvayconf.jpg')
solvayconf_img2 = cv2.imread('/content/drive/MyDrive/PCVK/Face Detection/solvayconf.jpg')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def convertToGRAY(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

faces_rects = faceCascade.detectMultiScale(jokowi_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(jokowi_img2, (x,y),(x+w,y+h),(0,255,0),5)

faces_rects = faceCascade.detectMultiScale(kartini_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(kartini_img2, (x,y),(x+w,y+h),(0,255,0),5)

faces_rects = faceCascade.detectMultiScale(kucing_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(kucing_img2, (x,y),(x+w,y+h),(0,255,0), 5)

faces_rects = faceCascade.detectMultiScale(mask_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(mask_img2, (x,y),(x+w,y+h),(0,255,0), 5)

faces_rects = faceCascade.detectMultiScale(jordan_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(jordan_img2, (x,y),(x+w,y+h),(0,255,0), 3)

faces_rects = faceCascade.detectMultiScale(solvayconf_img2, scaleFactor = 1.2, minNeighbors = 5);
for (x,y,w,h) in faces_rects:
     cv2.rectangle(solvayconf_img2, (x,y),(x+w,y+h),(0,255,0), 3)
```

```python
f, axarr = plt.subplots(6, 2, figsize = (15, 10))
plt.subplot(321), plt.imshow(convertToRGB(jokowi_img1))
plt.subplot(322), plt.imshow(convertToRGB(jokowi_img2))

plt.subplot(323), plt.imshow(convertToRGB(kartini_img1))
plt.subplot(324), plt.imshow(convertToRGB(kartini_img2))

plt.subplot(325), plt.imshow(convertToRGB(kucing_img1), cmap='gray')
plt.subplot(326), plt.imshow(convertToRGB(kucing_img2), cmap = 'gray')

f, axarr = plt.subplots(6, 2, figsize = (15, 10))
plt.subplot(321), plt.imshow(convertToRGB(mask_img1), cmap='gray')
plt.subplot(322), plt.imshow(convertToRGB(mask_img2), cmap='gray')

plt.subplot(323), plt.imshow(convertToRGB(jordan_img1), cmap='gray')
plt.subplot(324), plt.imshow(convertToRGB(jordan_img2), cmap='gray')

plt.subplot(325), plt.imshow(convertToRGB(solvayconf_img1), cmap='gray')
plt.subplot(326), plt.imshow(convertToRGB(solvayconf_img2), cmap='gray')
```

```python
def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality){
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)

  data = eval_js('takePhoto({})'.format(quality))
  img = js_to_image(data)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  print(gray.shape)
  faces = face_cascade.detectMultiScale(gray)
  for(x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255, 0, 0),2)

  cv2.imwrite(filename, img)

  return filename
```
