```python
# Baca gambar KTP atau SIM dari penyimpanan lokal
gambar_identitas = cv2.imread('KTP.jpeg')  # Ganti 'nama_file_gambar.jpg' dengan nama gambar Anda

# Menutup nama propinsi
x1_p = 600  # Koordinat x kiri atas
y1_p = 2   # Koordinat y kiri atas
x2_p = 200  # Koordinat x kanan bawah
y2_p = 30   # Koordinat y kanan bawah

# Menutup nama kabupaten
x1_k = 550  # Koordinat x kiri atas
y1_k = 55   # Koordinat y kiri atas
x2_k = 250  # Koordinat x kanan bawah
y2_k = 25   # Koordinat y kanan bawah

# Tutup bagian nama propinsi dengan warna merah
cv2.rectangle(gambar_identitas, (x1_p, y1_p), (x2_p, y2_p), [0,0,255], -1)

# Tutup bagian nama kabupaten dengan warna putih
cv2.rectangle(gambar_identitas, (x1_k, y1_k), (x2_k, y2_k), [255,255,255], -1)

# Konversi gambar dari BGR ke RGB untuk ditampilkan dengan Matplotlib
gambar_rgb = cv2.cvtColor(gambar_identitas, cv2.COLOR_BGR2RGB)

# Tampilkan gambar dengan Matplotlib
plt.imshow(gambar_rgb)
plt.axis('off')  # Matikan sumbu x dan y
plt.show()
```
